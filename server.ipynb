```python
while True:
    x=s.recvfrom(100000000)
    client=x[1][0]
    data=x[0]
    print(data)
    data=pickle.loads(data)
    print(type(data))
    data = cv2.imdecode(data, cv2.IMREAD_COLOR)
    cv2.imshow('videochat', data)
    if cv2.waitKey(10) == 13:
        break
cv2.destroyAllWindows()
```
